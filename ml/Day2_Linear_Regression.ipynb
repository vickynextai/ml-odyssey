## 1) What is Linear Regression?
Linear Regression is a supervised learning method used to predict **continuous values** by fitting a straight (or linear) relationship between input features (X) and target (y). The model learns coefficients that best explain how features influence the target.


## 2) Math (Core Ideas)

### Model
\[
\hat{y} = \beta_0 + \beta_1 x_1 + \cdots + \beta_p x_p
\]

### Loss (Mean Squared Error)
\[
\text{MSE} = \frac{1}{n}\sum_{i=1}^{n} (y_i - \hat{y}_i)^2
\]

### Closed-form (Normal Equation)
\[
\boldsymbol{\beta} = (X^\top X)^{-1} X^\top y
\]

### Gradient Descent (idea)
Iteratively update parameters to reduce MSE:
\[
\beta_j \leftarrow \beta_j - \eta \cdot \frac{\partial \text{MSE}}{\partial \beta_j}
\]


## 3) Simple Linear Regression (1 feature) â€“ Demo
Predict price from house size (toy data).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Toy dataset
X = np.array([500, 750, 1000, 1250, 1500]).reshape(-1, 1)  # size (sq ft)
y = np.array([50, 75, 100, 125, 150])  # price (lakhs)

# Train
lr = LinearRegression()
lr.fit(X, y)

# Predict on training points (for line)
y_pred = lr.predict(X)

print("Coefficient (slope):", lr.coef_[0])
print("Intercept:", lr.intercept_)
print("Predicted price for 1200 sq.ft:", lr.predict([[1200]])[0])

# Plot
plt.scatter(X, y, label="Data")
plt.plot(X, y_pred, label="Best-fit line")
plt.xlabel("House Size (sq ft)")
plt.ylabel("Price (lakhs)")
plt.legend()
plt.show()
```
